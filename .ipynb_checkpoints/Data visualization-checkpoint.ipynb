```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1)
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.hist(bins=10, figsize=(7,9))
```

```python
fig_1 = sns.FacetGrid(df, col='Sex', row='Survived', margin_titles=True)
fig_1.map(plt.hist, "Age", color='purple')
fig_1.fig.suptitle('Survival by Age and Sex')
plt.subplots_adjust(top=0.8)
```

```python
fig_2 = sns.FacetGrid(df, col='Sex', hue='Survived', margin_titles=True, palette={1:"blue", 0:"red"}, hue_kws=dict(marker=["v", "^"]))
fig_2.map(plt.scatter,"Age", "Fare", edgecolor="w").add_legend()
fig_2.fig.suptitle('Survival by Gender, Age, Fare')
plt.subplots_adjust(top=0.8)
```

```python
fig_3 = sns.FacetGrid(df, hue="Survived", col = "Pclass", margin_titles=True, palette={1:"blue", 0:"red"}, hue_kws=dict(marker=["v", "^"]))
fig_3.map(plt.scatter, "Age", "Fare", edgecolor="w").add_legend()
plt.subplots_adjust(top=0.8)
fig_3.fig.suptitle('Survival by Ticket Class, Fare and Age')
```

```python
fig_4 = sns.FacetGrid(df, col='Pclass', row='Survived', margin_titles=True)
fig_4.map(plt.hist, 'Fare', color='purple')
fig_4.fig.suptitle('Survival by Ticket Class and Fare')
plt.subplots_adjust(top=0.8)
```

```python
fig_5 = df.Embarked.value_counts().plot(kind='bar')
plt.title("Passengers per Port of Embarkation")
```

```python
sns.barplot(x='Embarked', y='Survived', hue='Sex', data=df)
```

```python
fig_6 = sns.factorplot(data= df, x='Embarked', y='Survived', col='Pclass', kind= 'bar', ci= None, aspect=0.6)
fig_6.fig.suptitle('Survival rate by Port of Embarkation and Ticket Class')
plt.subplots_adjust(top=0.8)
```

```python
fig_7 = sns.factorplot(data= df, x='Embarked', y='Survived', color='r')
fig_7.fig.suptitle('Survival rate by Port of Embarkation')
plt.subplots_adjust(top=0.8)
```

```python
fig_8 = sns.factorplot(data=df, x='Sex', y='Survived',col='Pclass', kind='bar', ci=None, aspect=0.6)
(fig_8.set_axis_labels("", "Survival Rate")
    .set_xticklabels(["Men", "Women"])
    .set(ylim=(0, 1))
    .despine(left=True))  
plt.subplots_adjust(top=0.8)
fig_8.fig.suptitle('How many Men and Women Survived by Passenger Class');
```

```python
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=df, markers=['*', 'o'])
```

```python
fig_9 = sns.boxplot(x='Survived', y='Age', data=df)
fig_9 = sns.stripplot(data=df, x='Survived', y='Age', jitter=True)
plt.title("Survival by Age")
```

```python
df.Age[df.Pclass == 1].plot(kind='kde')
df.Age[df.Pclass == 2].plot(kind='kde')
df.Age[df.Pclass == 3].plot(kind='kde')

plt.xlabel('Age')
plt.title("Age Distribution within classes")
plt.legend(('1st Class', '2nd Class', '3rd Class'), loc='best')
```

```python
sns.factorplot(x="Age", y="Embarked",
                    hue="Sex", row="Pclass",
                    data=df[df.Embarked.notnull()],
                    orient="h", size=2, aspect=3.5, 
                   palette={'male':"purple", 'female':"blue"},
                    kind="violin", split=True, cut=0, bw=.2);
```
